```python
import pandas as pd
import numpy as np
from scipy import stats
from scipy.stats import chi2_contingency
```

Question 1 - Cutlets

```python
#Analysis from a problem:
#H0: No difference in diameters, H1: Diameters are diffrent
#2 sample and 2 tail test
#performing t-test
#Î± = 0.05
```

```python
cutlet=pd.read_csv("/content/Cutlets.csv")
cutlet
```

```python
p_value=stats.ttest_ind(cutlet['Unit A'],cutlet['Unit B'])
p_value
```

```python
print("P value:",round(p_value[1],3))
```

```python
#Î± = 0.05
#p>Î± - Accept null hypothesis
#conclusion: There is no significant difference in the diameter of the cutlet between two units.
```

Question 2-LabTAT

```python
#Analysis from a problem:
#H0: No difference in TAT , H1: Diffrence in TAT
#4 samples
#performing Anova test
#Î± = 0.05
```

```python
lab=pd.read_csv("/content/LabTAT.csv")
lab
```

```python
p_value=stats.f_oneway(lab['Laboratory 1'],lab['Laboratory 2'],lab['Laboratory 3'],lab['Laboratory 4'])
p_value
```

```python
#Î± = 0.05 and p>Î± : Accept the null hypothesis
#conclusion: There is difference in TAT
```
