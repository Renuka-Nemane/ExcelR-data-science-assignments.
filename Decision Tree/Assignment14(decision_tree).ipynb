```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import  DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import classification_report
from sklearn import preprocessing
import warnings
warnings.filterwarnings('ignore')
```

**Fraud Check Dataset**

```python
fraud = pd.read_csv('/content/Fraud_check.csv')
fraud
```

```python
fraud1 = fraud.drop(['City.Population'], axis = 1)
fraud1
```

```python
fraud1.info()
```

```python
y1=np.empty(600, dtype=object)
i=0

for value in fraud1['Taxable.Income']:
    if value<=30000:
        y1[i]='Risky'
    else:
        y1[i]='Good'
    i=i+1
```

```python
y1
```

```python
t1=pd.DataFrame(y1,columns=['Target'])
t1
```

```python
fraud2 = pd.concat([fraud1,t1],axis=1)
fraud2.head()
```

```python
fraud2.info()
```

```python
fraud2.corr()
```

```python
fraud2.groupby(['Undergrad', 'Marital.Status' ,'Urban']).count()
```

```python
label_encoder = preprocessing.LabelEncoder()
fraud2['Undergrad']= label_encoder.fit_transform(fraud['Undergrad'])
fraud2['Marital.Status']= label_encoder.fit_transform(fraud['Marital.Status'])
fraud2['Urban']= label_encoder.fit_transform(fraud['Urban'])
fraud2['Target'] = fraud2['Target'].astype(str)
```

```python
from sklearn.tree import DecisionTreeRegressor
array = fraud2.values
X = array[:,0:5]
y = array[:,0]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=21)
```
